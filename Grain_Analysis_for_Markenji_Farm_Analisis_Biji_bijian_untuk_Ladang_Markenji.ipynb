## Grain Analysis for Markenji Farm / Analisis Biji-bijian untuk Ladang Markenji

### Latar Belakang

Markenji Farm adalah perusahaan yang berfokus pada bidang pertanian yang mana hasil dari pertanianya sebagai berikut; Corn, Oat, Cereals & Grins. Letak Markenji Farm sendiri ada pada benua eropa. Dikarenakan pembelian hasil pertanian melalui obligasi dimana harga yang sekarang dibeli tidak dapat dirubah oleh Markenji Farm karena terikat kontrak. Sehingga ini menjadi tantangan bagi Data Analis di Markenji Farm untuk dapat memberikan insight agar pada musim apa hasil jenis panen mendapatkan penawaran terbaik.

Musim di Eropa mengikuti pola umum yang sama seperti di belahan bumi utara lainnya, dengan empat musim utama: musim semi, musim panas, musim gugur, dan musim dingin. Namun, perubahan musimnya bisa sedikit berbeda tergantung pada lokasi geografis di Eropa. Di bawah ini adalah pembagian musim di Eropa berdasarkan bulan:

`Musim Semi`:

Maret, April, Mei: Ini adalah bulan-bulan awal musim semi. Suhu mulai naik, salju meleleh, dan tumbuhan mulai mekar. Musim semi di Eropa biasanya dimulai dari selatan ke utara.

`Musim Panas`:

Juni, Juli, Agustus: Musim panas di Eropa adalah saat cuaca umumnya hangat hingga panas. Suhu puncak musim panas biasanya terjadi pada bulan Juli. Selama musim panas, banyak orang berlibur di pantai, dan festival serta kegiatan luar ruangan menjadi populer.

`Musim Gugur`:

September, Oktober, November: Pada musim gugur, suhu mulai turun, dan daun-daun pohon berubah warna sebelum gugur. Ini adalah musim yang indah dengan panorama alam yang menakjubkan.

`Musim Dingin`:

Desember, Januari, Februari: Musim dingin di Eropa adalah saat suhu merosot drastis, dan salju seringkali turun di sebagian besar wilayah Eropa. Musim dingin dapat sangat dingin terutama di bagian utara Eropa seperti Skandinavia.
Namun, perlu diingat bahwa ini adalah pembagian musim yang umum, dan perubahan musim bisa sedikit berbeda di berbagai negara dan wilayah di Eropa. Beberapa negara di selatan Eropa, seperti Spanyol, Italia, dan Yunani, cenderung memiliki musim panas yang lebih panjang dan musim dingin yang lebih ringan dibandingkan dengan negara-negara di utara seperti Swedia, Norwegia, dan Rusia.

Biji-bijian Eropa, seperti gandum, jagung, barley, dan lainnya, diekspor ke berbagai negara di seluruh dunia. Beberapa benua yang umumnya merupakan tujuan utama untuk ekspor biji-bijian Eropa termasuk:

  `Asia`:
  Asia adalah salah satu pasar terbesar untuk biji-bijian Eropa. Negara-negara seperti China, India, dan Jepang adalah konsumen utama biji-bijian Eropa.

  `Amerika Utara`:
  Amerika Serikat dan Kanada adalah dua negara utama di Amerika Utara yang mengimpor biji-bijian Eropa seperti gandum dan jagung.

  `Afrika`:
  Beberapa negara di Afrika, terutama negara-negara dengan populasi besar dan ekonomi berkembang seperti Mesir, Nigeria, dan Aljazair, adalah tujuan utama bagi biji-bijian Eropa.

  `Amerika Selatan`:
  Beberapa negara di Amerika Selatan, seperti Brasil, juga mengimpor biji-bijian Eropa.

### Membuat Column Kondisi 1 dan 2

Kondisi Untung 1 diberikan jika harga penutupan lebih tinggi dari pada harga terendah maka akan diberi label 1 dengan kondisi untung. Sedangkan jika label 0 maka kondisi rugi.

Kondisi Untung 2 diberikan jika harga penutupan lebih tinggi dari pada harga tertinggi maka akan diberi label 1 dengan kondisi untung besar. Sedangkan jika label 0 maka kondisi untung.

```python
df_1['Kondisi Untung 1'] = np.where(df_1['close'] >= df_1['low'], 1, 0)
```

```python
df_1['Kondisi Untung 1'].value_counts()
```

```python
df_1['Kondisi Untung 2'] = np.where((df_1['close'] >= df_1['high']) & (df_1['close'] >= df_1['low']), 1, 0)
df_1['Kondisi Untung 2'].value_counts()
```

```python
df_1.head()
```

### Membuat Column Total Penjualan

Total Penjualan adalah hasil perkalian antara volume dan penutupan harga yang diambil sebagai keuntungan penjualan secara kotor atau gross.

```python
df_1["Total Penjualan"] = df_1["close"]*df_1["volume"]
df_1.head()
```

### Merubah Tipe Data

Merubah tipe data date dari object ke date.

```python
df_1['Date'] = pd.to_datetime(df_1['date'])
df_1.head()
```

```python
# Melakukan drop column date lama
df_1 = df_1.drop('date', axis=1)
```

Membuat kolum Year dari kolum Date

```python
df_1['Year'] = df_1['Date'].dt.year
```

Membuat kolum Month dari kolum Date

```python
df_1['Month'] = df_1['Date'].dt.month
df_1.head()
```

```python
df_1.info()
```

### Data Visualization & EDA

#### Normalitas

```python
df_gambar = ['open', 'high', 'low', 'close', 'volume',
       'Kondisi Untung 1', 'Kondisi Untung 2', 'Total Penjualan']
```

```python
len(df_gambar)
```

```python
#plot histogram untuk tiap fitur numerikal
count=0
fig = plt.figure(figsize=(15, 30))
for feature in df_gambar:
    count +=1
    ax= plt.subplot(5,2,count)
    sns.histplot(data=df_1, x=feature, color="dodgerblue")
    plt.title("Plot for {}".format(feature))
    fig.tight_layout()
plt.show()
```

Data yang ada pada setiap kolum pada tabel terindikasi tidak terdistribusi secara normal. Sehingga akan dilakukan statistical test untuk menguji distribusinya.

#### **Statistical Test Method - Shapiro Wilk**

```python
# melakukan uji normalitas dengan metode Shapiro Wilk
# import library
from scipy import stats

print('''
H0: Data Terbukti Terdistribusi Normal
H1: Data Tidak Terbukti Terdistribusi Normal
apabila p_val >= 0.05 Terima H0, Tolak H1
apabila p_val < 0.05 Tolak H0, Terima H1
''')
#fungsi normalitas
for feature in df_gambar:
  stat, p_val = stats.shapiro(df_1[feature].dropna())
  if p_val >= 0.05:
    print("{}\t Terbukti Terdistribusi Normal\n".format(feature))

  else:
    print("{}\t Tidak Terbukti Terdistribusi Normal\n".format(feature))
```

Setelah dilakukan statistical test ternyata data pada setiap kolum terbukti tidak terdistribusi secara normal.

### Outliers

```python
#boxplot untuk visualisasi outlier
count=0
fig = plt.figure(figsize=(15, 30))
for feature in df_gambar:
    count +=1
    ax= plt.subplot(10,2,count)
    sns.boxplot(data=df_1,x=feature, color="dodgerblue")
    plt.title("Plot for {}".format(feature))
    fig.tight_layout()
plt.show()
```

Ditemukan outlier pada kolum pada setiap kolum pada tabel kecuali pada Kondisi Untuk 1 dan 2. Sehingga data yang kita miliki, memiliki anomali yang mana dapat kita temukan pettern yang unik.

#### Apakah Ticker, Commodity, Open, Close memiliki korelasi satu sama lain ?

### Correlation Heatmap

```python
#analisa korelasi data dengan metode spearman
biji_corr = df_1.corr(method="spearman", numeric_only=True)
biji_corr
```

Menggunakan methode spearman karena distribusi data tidak normal.

```python
#peta heatmap untuk mempermudah melihat korelasi antar fitur
plt.figure(figsize = (15,7))

trimask = np.triu(np.ones_like(biji_corr))
sns.heatmap(data=biji_corr, annot=True, cmap="Blues", mask=trimask|np.abs(biji_corr<=0.4))
plt.title("Heatmap Korelasi Grain")
plt.show()
```

Pada Correlation Heatmap terlihat bahwa open, high, dan low memiliki korelasi yang tinggi sampe sama dengan 1. Sedangkan korelasi tinggi lainya terletak pada total penjualan dengan volume. Korelasi rendah terdapat pada Year dengan volume dan total penjualan.
